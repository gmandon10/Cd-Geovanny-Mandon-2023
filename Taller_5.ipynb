```python
import numpy as np
import matplotlib.pyplot as plt

lado_cuadricula = 1.0
num_puntos = 1000
distancia_entre_puntos = 1/(num_puntos-1)
coordenadas_x = []
coordenadas_y = []
suma = 0

for i in range(num_puntos):
    for j in range(num_puntos):
        x = i * distancia_entre_puntos
        y = j * distancia_entre_puntos
        coordenadas_x.append(x)
        coordenadas_y.append(y)
        distancia = np.sqrt(x**2 + y**2)

        if distancia <= 1:
            suma += 1

plt.scatter(coordenadas_x, coordenadas_y, s=10)
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.title('Puntos Equidistantes en una Cuadrícula de Lado 1')
plt.grid(True)
plt.show()
print("")
print("Suma de distancias menores o iguales a 1:", suma)
print("Estimación de π cuartos:", suma/num_puntos)
print("Estimación de π:", suma/(num_puntos**2)*4)
```

```python
import numpy as np
import matplotlib.pyplot as plt

lado_cuadricula = 1.0
num_puntos = 1000
coordenadas_x = []
coordenadas_y = []
suma = 0

for _ in range(num_puntos * num_puntos):
    x = np.random.uniform(0, lado_cuadricula)
    y = np.random.uniform(0, lado_cuadricula)
    coordenadas_x.append(x)
    coordenadas_y.append(y)

    distancia = np.sqrt(x**2 + y**2)

    if distancia <= 1:
        suma += 1

plt.scatter(coordenadas_x, coordenadas_y, s=10)
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.title('Puntos Aleatorios en una Cuadrícula de Lado 1')
plt.grid(True)
plt.show()
print("")
print("Suma de distancias menores o iguales a 1:", suma)
print("Estimación de π cuartos:", suma/(num_puntos**2))
print("Estimación de π:", suma/(num_puntos**2)*4)
```
