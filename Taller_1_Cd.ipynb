```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
id = [1,2,3,4,5,6,7,8]
area = [75,35,105,50,85,72,120,90]
precio = [200,120,210,155,300,240,245,250]
```

```python
area = np.array(area).reshape(-1, 1)
precio = np.array(precio).reshape(-1, 1)

regresion = linear_model.LinearRegression()
regresion.fit(area, precio)
area_pred = regresion.predict(area)

print('Coeficientes: \n', regresion.coef_)

print('Termino independiente: \n', regresion.intercept_)

print("Error cuadrado medio: %.2f" % mean_squared_error(precio, area_pred))

print('Varianza: %.2f' % r2_score(precio, area_pred))
```

```python
sns.regplot(x = area, y = precio)
```
